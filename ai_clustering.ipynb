```python
import re
def clean_name(name):
    if pd.isna(name):
        return name
    name = re.sub(r"\s+", " ", name)  # Remove extra spaces
    name = re.sub(r"[^\w\s\/]", "", name)  # Remove special characters
    name = name.strip()  # Trim spaces
    return name.lower().title()  # Capitalize appropriately

df['name'] = df['name'].apply(clean_name)
```

```python
df['packaging_form'] = df['packaging_form'].str.lower().str.strip()  # Convert to lowercase and remove spaces
```

```python
df['salts'] = df['salts'].str.lower().str.strip()  # Convert to lowercase and remove spaces
```

```python
def clean_quantity(value, name):
    if pd.isna(value):
        return np.nan

    value = str(value).lower()  # Convert to lowercase to standardize

    # Extract the numeric portion from the quantity string
    numeric_value = ''.join(filter(str.isdigit, value))


    if 'ml' in value:
        return f"{numeric_value} ml"
    elif 'tablet' in value or 'tablets' in value:
        return f"{numeric_value} tablet"
    elif 'syrup' in value or 'suspension' in value or 'liquid' in value:
        return f"{numeric_value} ml"
    return value

df['quantity'] = df.apply(lambda row: clean_quantity(row['quantity'], row['name']), axis=1)

print(df)
```

```python
def fill_packaging_form(row):
    if pd.isnull(row['packaging_form']):
        name = row['name'].lower()

        if 'tablet' in name:
            row['packaging_form'] = 'strip'
        elif 'bottle' in name or 'suspension' in name:
            row['packaging_form'] = 'bottle'
        elif 'infusion' in name:
            row['packaging_form'] = 'bottle'  # Assuming Infusion is in a bottle
        elif 'ampule' in name:
            row['packaging_form'] = 'ampule'
        else:
            row['packaging_form'] = 'other'  # Default case if no match is found

    return row

# Apply the function to fill 'packaging_form'
df = df.apply(fill_packaging_form, axis=1)

# Check the updated DataFrame
print(df)
```

```python
df = df.dropna(subset=["quantity"])
```

```python
mode_value = df["salts"].mode()[0]

# Fill NaN values in 'salts' column with the mode
df["salts"].fillna(mode_value, inplace=True)

# Display the updated DataFrame
print(df)
```

```python
df = df.drop(columns=["id", "source"])
```

```python
print(df)
```

```python
def add_unit(row):
    name_lower = row["name"].lower()  # Convert name to lowercase
    quantity_lower = row["quantity"].lower()  # Convert quantity to lowercase

    # Check if "tablet" or "tablets" is in the name and no unit is present in quantity
    if ("tablet" in name_lower or "tablets" in name_lower) and "tablet" not in quantity_lower and "ml" not in quantity_lower:
        return str(row["quantity"]) + " tablet"
    # Check if "ml" is missing and "tablet" is not in the name
    elif "tablet" not in name_lower and "ml" not in quantity_lower:
        return str(row["quantity"]) + " ml"
    else:
        return row["quantity"]

# Apply the function to the DataFrame
df["quantity"] = df.apply(add_unit, axis=1)

# Convert the 'name' column to lowercase
df["name"] = df["name"].str.lower()

# Display the updated DataFrame
print(df)
```
